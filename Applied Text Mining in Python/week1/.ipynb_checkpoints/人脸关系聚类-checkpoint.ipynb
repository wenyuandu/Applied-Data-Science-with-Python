```python
def female_number(faces):
    female_number = 0
    for face in faces:
        if face['gender'] == '女':
            female_number = female_number+1
    return female_number
```

```python
def male_number(faces):
    male_number = 0
    for face in faces:
        if face['gender'] == '男':
            male_number = male_number+1
    return male_number
```

```python
def age2age_number(faces, fromAge, toAge):
    age2age_number = 0
    for face in faces:
        if face['age'] > fromAge and face['age'] <= toAge:
            age2age_number = age2age_number+1
    return age2age_number
```

```python
import pandas as pd
import json



data = pd.read_table('/Users/wenyuan/文件/study/luface/人脸数据.txt',header=None, sep='\t', names=["url", "faces", "date"], quoting = 3)

data['faces'] = data['faces'].apply(lambda x:  json.loads(json.loads(x)))
data['face_number'] = data['faces'].apply(lambda x:  len(x))
data = data.loc[data['face_number'] > 1]



data['female_number'] = data['faces'].apply(lambda faces:  female_number(faces) )
data['male_number'] = data['faces'].apply(lambda faces:  male_number(faces) )
data['zero2ten_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 0, 10) )
data['ten2twenty_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 10, 20) )
data['twenty2thirty_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 20, 30) )
data['thirty2forty_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 30, 40) )
data['forty2fifty_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 40, 50) )
data['fifty2sixty_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 50, 60) )
data['sixty2senventy_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 60, 70) )
data['senventy2eighty_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 70, 80) )
data['eighty2ninety_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 80, 90) )
data['ninety2hundred_number'] = data['faces'].apply(lambda faces:  age2age_number(faces, 90, 100) )



data.head(50)

# data.to_csv('/Users/wenyuan/文件/study/luface/人脸图片聚合数据.txt',index=False,header=True)


```

```python
train = data[
    ['female_number', 'male_number',
     'zero2ten_number', 'ten2twenty_number', 'twenty2thirty_number', 'thirty2forty_number', 'forty2fifty_number', 
     'fifty2sixty_number', 'sixty2senventy_number', 'senventy2eighty_number', 'eighty2ninety_number', 'ninety2hundred_number', 
    ]
    ]

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt



tsne = TSNE(n_components=2, random_state=0) # first try without init='pca'
X_tsne = tsne.fit_transform(train) 
    
xtsne = pd.DataFrame(X_tsne)
 

```

```python
xtsne.info() 
```

```python
c_map = {0:'b', 1:'r'}
plt.figure(figsize = (10,9))
plt.scatter(X_tsne[:,0],X_tsne[:,1], alpha=0.5)
#plt.ylim(-10,10)
plt.show()
```

```python
tsne_bis = TSNE(n_components=2, random_state=0, init='pca')
X_tsne_bis = tsne_bis.fit_transform(train)
xtsne_b = pd.DataFrame(X_tsne_bis)
c_map = {0:'b', 1:'r'}
fig = plt.figure(figsize = (10,9))
plt.scatter(X_tsne_bis[:,0],X_tsne_bis[:,1] ,alpha=0.5)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=7)
# Compute cluster centers and predict cluster indices
X_clustered = kmeans.fit(train)

```

```python
df2=pd.DataFrame(X_clustered.cluster_centers_,columns=['female_number', 'male_number',
     'zero2ten_number', 'ten2twenty_number', 'twenty2thirty_number', 'thirty2forty_number', 'forty2fifty_number', 
     'fifty2sixty_number', 'sixty2senventy_number', 'senventy2eighty_number', 'eighty2ninety_number', 'ninety2hundred_number', 
    ])
pd.set_option('display.float_format', lambda x: '%.1f' % x)

print(df2)
```
